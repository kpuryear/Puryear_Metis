```python
import pandas as pd
from transit_periodogram import transit_periodogram
import numpy as np
import matplotlib.pyplot as plt
from lightkurve import KeplerTargetPixelFile
```

```python
def singlestarpipe(kepid):
    #string = str(kepid)
    tpf = KeplerTargetPixelFile.from_archive(kepid, quarter=1)  
    lc = tpf.to_lightcurve(aperture_mask=tpf.pipeline_mask)
    flat, trend = lc.flatten(window_length=301, return_trend=True)
    df = flat.to_pandas()
    df["KeplerID"] = kepid
    return df

def attachpandastogether(df, ds):
    result = df.append(ds)
    return result
```
