```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# upload data from locally-saved CSV file

df = pd.DataFrame.from_csv('/Users/kaitlin/ds/metis/metisgh/Puryear_Metis/Project3/NoShows.csv', 
                      header=0, 
                      sep=',', 
                      index_col=0, 
                      parse_dates=True, 
                      encoding=None, 
                      tupleize_cols=None, 
                      infer_datetime_format=False)
```

```python
# raw data comes in looking like...

df.head()
```

```python
# fix typos in column headings...

df = df.rename(index=str, columns={"SMS_received": "Text_Sent", 
                              "Hipertension": "Hypertension", 
                              "Handcap": "Handicap",
                                  "Neighbourhood": "neighborhood"})
df.head()
```

```python
# boolean True or False for response variable..

df["Showed_up"] = (df["No-show"] == "No")
```

```python
# Make all values in X booleans!

df["woman"] = (df["Gender"] == 'F')
df["Scholarship"] = (df["Scholarship"] == 1)
df["Hypertension"] = (df["Hypertension"] == 1)
df["Diabetes"] = (df["Diabetes"] == 1)
df["Alcoholism"] = (df["Alcoholism"] == 1)
df["Handicap"] = (df["Handicap"] == 1)
df["Text_Sent"] = (df["Text_Sent"] == 1)
```

```python
# create datetime objects for appointment_today and scheduled_today...

df["appointment_today"] = [datetime.strptime(x,'%Y-%m-%dT%H:%M:%SZ') for x in df["AppointmentDay"]]

df["scheduled_today"] = [datetime.strptime(x,'%Y-%m-%dT%H:%M:%SZ') for x in df["ScheduledDay"]]
```

```python
# create new variable "delta" which is the difference between the two datetimes...

df["delta"] = [((x-y).days)+1 for x,y in zip(df['appointment_today'], df["scheduled_today"])]
```

```python
df.head()
```

```python
# get rid of redundant variables

df = df.drop(['AppointmentDay',
              'appointment_today',
              'scheduled_today',
              'Gender',
              'ScheduledDay', 
              'AppointmentID',
              'No-show'], axis=1)
```

```python
# get rid of patient ID, and create generalized index values...

df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
# drop NaNs...

df2 = pd.DataFrame.dropna(df)
```

```python
# neighborhoods obfuscate the data too much...

df2 = df2.drop('neighborhood', axis=1)
```

```python
df2.head()
```

```python
# export the cleaned data as a csv file
# get it in good form to upload into SQL database

df2.to_csv('noshowclean_n.csv', encoding='utf-8', index=True)
```
